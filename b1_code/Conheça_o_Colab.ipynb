```
import math


def equacao2grau(a,b,c):
  delta = b**2 - 4 * a * c
  if delta  < 0:
    return"Sem raiz"
  x1 = (-b + math.sqrt(delta)) / (2*a)
  x2 = (-b - math.sqrt(delta)) / (2*a)
  return x1, x2
```

```
print(equacao2grau(1, -3, 2))
```
